**1. Import Library**

```python
import matplotlib.pyplot as plt
```

**2. Buat Dataset**

```python
suhu = [20, 21, 22, 23, 24, 25, 26, 27, 28]
penjualan = [50, 60, 70, 80, 90, 100, 110, 120, 130]
```

**3. Basic Scatter Plot**

```python
plt.scatter(suhu, penjualan)
plt.title('Hubungan Penjualan Es Krim dgn Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.show()
```

**4. Kustomisasi Scatter PLot**

```python
plt.scatter(suhu, penjualan, color='red', edgecolor='black', s=100, alpha=0.7)
plt.title('Hubungan Penjualan Es Krim dgn Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.grid(True)
plt.show()
```

**5. Multiple Scatter Plot**

```python
#dataset suhu
suhu = [20, 21, 22, 23, 24, 25, 26, 27, 28]

#dataset penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [40, 50, 60, 70, 80, 90, 100, 110, 120]

plt.scatter(suhu, penjualan_weekdays, color='skyblue', label='Hari Kerja', s=80)
plt.scatter(suhu, penjualan_weekend, color='pink', label='Akhir Pekan', s=80)
plt.title('Hubungan Penjualan Es Krim dgn Suhu')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.legend()
plt.show()
```

**6. Melihat Hubungan Data dgn beberapa Variabel**

```python
#dataset
suhu = [20, 21, 22, 23, 24, 25, 26, 27, 28]
penjualan = [50, 60, 70, 80, 90, 100, 110, 120, 130]

#dataset tambahan
kelembapan = [60, 65, 70, 75, 80, 85, 90, 95, 100]
ukuran_titik = [100, 120, 140, 160, 180, 200, 220, 240, 260]

#kustomisasi scatter plot
plt.scatter(suhu, penjualan, c=kelembapan, cmap='coolwarm', s=ukuran_titik, alpha=0.6)
plt.title('Hubungan Penjualan, suhu, dan kelembapan')
plt.xlabel('Suhu')
plt.ylabel('Penjualan Es Krim')
plt.colorbar(label='Kelembapan (%)')
plt.show()
```

**7. 3D Scatter Plot**

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#dataset suhu
suhu = [20, 21, 22, 23, 24, 25, 26, 27, 28]

#dataset penjualan
penjualan_weekdays = [50, 60, 70, 80, 90, 100, 110, 120, 130]
penjualan_weekend = [40, 50, 60, 70, 80, 90, 100, 110, 120]

#3D Scatter Plot
fig=plt.figure(figsize=(15, 13))
ax=fig.add_subplot(111, projection='3d')

ax.scatter(suhu, penjualan_weekdays, penjualan_weekend, color='purple', label='Data Penjualan', s=80)
ax.set_title('Hubungan Penjualan Es Krim dan suhu (3D ver)')
ax.set_xlabel('Suhu')
ax.set_ylabel('Penjualan Weekdays')
ax.set_zlabel('Penjualan Weekend')
ax.legend()
plt.show()
```
